# IMDb Data

```python
f = open("../Files/imdb_1.json")
imdb_json = json.load(f)

#imdb_1
```

```python
# Create dataframe with IMDb data
imdb_df = pd.json_normalize(imdb_json)
imdb_df
```

```python
imdb_df.info()
```

```python
# Drop irrelevant columns
imdb_df = imdb_df.drop(columns=["Rated", "Released", "Genre", "Director", "Writer", "Actors", "Language", "Country", "Poster", "Ratings", "Type", "DVD", "BoxOffice", "Production", "Website", "Response"])
```

```python
# Drop rows without IMDb ratings
imdb_df = imdb_df.drop(imdb_df.loc[imdb_df["imdbRating"] == "N/A"].index)
imdb_df.head(2)
```

```python
# Drop rows without runtime
imdb_df = imdb_df.drop(imdb_df.loc[imdb_df["Runtime"] == "N/A"].index)
imdb_df.head(2)
```

```python
# Replace N/A in Metascore with 0
imdb_df["Metascore"] = imdb_df["Metascore"].str.replace("N/A", "0")
imdb_df.head(2)
```

```python
# Remove comma from IMDb votes
imdb_df["imdbVotes"] = imdb_df["imdbVotes"].str.replace(",", "")
imdb_df.head(2)
```

```python
# Remove "min" from runtime
imdb_df["Runtime"] = imdb_df["Runtime"].str.strip(" min")
imdb_df.head(2)
```

```python
imdb_df = imdb_df.astype({"Year": "int", "Runtime": "int", "Metascore": "int", "imdbRating": "float", "imdbVotes": "int"})
```

```python
imdb_df.info()
```

```python
# Create new columns for Oscars nominations and wins
imdb_df["OscarNominations"] = imdb_df["Awards"]
imdb_df["OscarWins"] = imdb_df["Awards"]
```

```python
# Extract Oscar nominations
imdb_df["OscarNominations"] = imdb_df.Awards.str.extract(r"((Nominated for)\s\d\s(Oscar))")

# Extract Oscar wins
imdb_df["OscarWins"] = imdb_df.Awards.str.extract(r"((Won)\s\d\s(Oscar))")
```

```python
# Extract number of Oscar nominations
imdb_df["OscarNominations"] = imdb_df.OscarNominations.str.extract(r"(\d)")

# Extract number of Oscar wins
imdb_df["OscarWins"] = imdb_df.OscarWins.str.extract(r"(\d)")
```

```python
# Fill nulls with 0
imdb_df["OscarNominations"] = imdb_df["OscarNominations"].fillna(0)
imdb_df["OscarWins"] = imdb_df["OscarWins"].fillna(0)
```

```python
# Drop awards column
imdb_df = imdb_df.drop(columns=["Awards"])
imdb_df
```

```python
# Store Rotten Tomatoes ratings in list
ratings_rotten = []
for title in imdb_json:
    for rating in title["Ratings"]:
        if rating["Source"] == "Rotten Tomatoes":
            rating_rotten = rating["Value"]
    ratings_rotten.append([title["imdbID"], rating_rotten])
        
ratings_rotten[:10]
```

```python
# Create dataframe of Rotten Tomatoes ratings
ratings_rotten_df = pd.DataFrame(ratings_rotten, columns=["imdbID", "RottenTomatoesRating"])
```

```python
# Remove % signs from Rotten Tomatoes ratings
ratings_rotten_df["RottenTomatoesRating"] = ratings_rotten_df["RottenTomatoesRating"].str.strip("%")
ratings_rotten_df
```

```python
ratings_rotten_df.info()
```

```python
# Merge Rotten Tomatoes ratings into IMDb dataframe
imdb_df = imdb_df.merge(ratings_rotten_df, on="imdbID")
imdb_df
```

# Gender from TMDb

```python
# Function to store gender details in dataframe
def gender_json_to_df(json, column_1, column_2):
    new_list = []
    for key, results in json.items():
        for person in results["person_results"]:
            new_list.append([key, person["id"], person["gender"]])
    
    new_df = pd.DataFrame(new_list, columns=[column_1, column_2, "Gender"])
    return new_df              
```

# Other Details from TMDb

```python
# Import TMBd movie details
f = open("../Files/tmdb_2.json")
tmdb_json = json.load(f)

#tmdb_json
```

```python
# Function to store details in dataframe
def tmdb_json_to_df(tmdb_json, key_name, value_name, new_column_name):
    new_list = [] 
    for tmdb_id in tmdb_json:
        temp_list = [categories for categories in tmdb_id[key_name]]
        temp_flat = [category[value_name] for category in temp_list]
        for item in temp_flat:
            new_list.append([tmdb_id["imdb_id"], tmdb_id["id"], item])
    
    new_df = pd.DataFrame(new_list, columns=["imdbID", "tmdbID", new_column_name])
    return new_df
```
