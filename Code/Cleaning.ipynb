# Libraries

```python
import json
import csv
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

# IMDb Data

```python
f = open("../Files/imdb_1.json")
imdb_json = json.load(f)
#imdb_1
```

```python
# Create dataframe with IMDb data
imdb_df = pd.json_normalize(imdb_json)
imdb_df
```

```python
imdb_df.info()
```

```python
# Drop irrelevant columns
imdb_df = imdb_df.drop(columns=["Released", "Writer", "Poster", "Ratings", "Type", "DVD", "BoxOffice", "Website", "Response"])
```

```python
# Import IMDb movie IDs (filtered for independent movies)
with open("../Files/imdb_ids_filtered.csv", newline="") as f:
    reader = csv.reader(f)
    ls = list(reader)
    imdb_ids_filtered = [item for items in ls for item in items]

imdb_ids_filtered[:10]
```

```python
# Create filtered dataframe by independent movies
imdb_df_filtered = imdb_df.loc[imdb_df["imdbID"].isin(imdb_ids_filtered)]
```

```python
# Create new columns for Oscars nominations and wins
imdb_df_filtered["OscarNominations"] = imdb_df_filtered["Awards"]
imdb_df_filtered["OscarWins"] = imdb_df_filtered["Awards"]
```

```python
# Extract Oscar nominations
imdb_df_filtered["OscarNominations"] = imdb_df_filtered.Awards.str.extract(r"((Nominated for)\s\d\s(Oscar))")

# Extract Oscar wins
imdb_df_filtered["OscarWins"] = imdb_df_filtered.Awards.str.extract(r"((Won)\s\d\s(Oscar))")
```

```python
# Extract number of Oscar nominations
imdb_df_filtered["OscarNominations"] = imdb_df_filtered.OscarNominations.str.extract(r"(\d)")

# Extract number of Oscar wins
imdb_df_filtered["OscarWins"] = imdb_df_filtered.OscarWins.str.extract(r"(\d)")
```

```python
# Fill nulls with 0
imdb_df_filtered["OscarNominations"] = imdb_df_filtered["OscarNominations"].fillna(0)
imdb_df_filtered["OscarWins"] = imdb_df_filtered["OscarWins"].fillna(0)
```

```python
# Drop awards column
imdb_df_filtered = imdb_df_filtered.drop(columns=["Awards"])
imdb_df_filtered
```

```python
# Store Rotten Tomatoes ratings in list
ratings_rotten = []
for title in imdb_json:
    for rating in title["Ratings"]:
        if rating["Source"] == "Rotten Tomatoes":
            rating_rotten = rating["Value"]
    ratings_rotten.append([title["imdbID"], rating_rotten])
        
ratings_rotten[:10]
```

```python
# Create dataframe of Rotten Tomatoes ratings
ratings_rotten_df = pd.DataFrame(ratings_rotten, columns=["imdbID", "RottenTomatoesRating"])
```

```python
# Remove % signs
ratings_rotten_df["RottenTomatoesRating"] = ratings_rotten_df["RottenTomatoesRating"].str.strip("%")
ratings_rotten_df
```

```python
ratings_rotten_df.info()
```

# IMDb Keywords

```python
f = open("../Files/imdb_keywords.json")
imdb_keywords_json = json.load(f)
#imdb_keywords_json
```

```python
# Create list of IMDb IDs and keywords
imdb_keywords = [] 
for imdb_id, value in imdb_keywords_json.items():
    keywords = value["Keywords"]
    keywords_list = [keyword["keyword"] for keyword in keywords]
    for item in keywords_list:
        imdb_keywords.append([imdb_id, item])

imdb_keywords[:20]
```

```python
# Create dataframe of IMDb keywords
imdb_keywords_df = pd.DataFrame(imdb_keywords, columns=["imdbID", "imdbKeyword"])
imdb_keywords_df
```

# TMDb Keywords

```python
f = open("../Files/tmdb_keywords.json")
tmdb_keywords_json = json.load(f)
#tmdb_keywords_json
```

```python
# Create list of TMDb IDs and keywords
tmdb_keywords = [] 
for tmdb_id in tmdb_keywords_json:
    keywords_list = [keywords for keywords in tmdb_id["keywords"]]
    keywords_flat = [keyword["name"] for keyword in keywords_list]
    for item in keywords_flat:
        tmdb_keywords.append([tmdb_id["id"], item])

tmdb_keywords[:20]
```

```python
# Create dataframe of TMDb keywords
tmdb_keywords_df = pd.DataFrame(tmdb_keywords, columns=["tmdbID", "tmdbKeyword"])
tmdb_keywords_df
```

# Gender of Directors

```python
# Import director details
f = open("../Files/tmdb_directors.json")
directors_json = json.load(f)

len(directors_json)
```

```python
# Create list of director IMDb and TMDb IDs, names and gender
gender_directors = []
for director, results in directors_json.items():
    for person in results["person_results"]:
        gender_directors.append([director, person["id"], person["name"], person["gender"]])

gender_directors[-1:]
```

```python
len(gender_directors)
```

```python
gender_directors_df = pd.DataFrame(gender_directors, columns=["imdbDirectorID", "tmdbDirectorID", "tmdbDirectorName", "Gender"])
gender_directors_df
```

```python
# Import IMDb director details
directors_df = pd.read_csv("../Files/imdb_directors.csv")
directors_df
```

```python
directors_df["imdbDirectorID"].value_counts()
```

```python
# Merge with gender
directors_df = directors_df.merge(gender_directors_df, how="left", on="imdbDirectorID")
```

```python
directors_df.info()
```

```python
directors_df = directors_df.dropna()
```

```python
directors_df = directors_df.astype({"Gender": "int", "tmdbDirectorID": "int"})
```

```python
directors_df.head()
```

```python
directors_df["Gender"].value_counts()
```
